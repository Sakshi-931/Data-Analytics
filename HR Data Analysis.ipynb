 - Assign his category to the Culumns.

```python
categorical_cols = ['Attrition','BusinessTravel','Department','Gender','JobRole','MaritalStatus']
for col in categorical_cols:
    df[col] = df[col].astype('category')
```

- Count plot used to Find atrribution Distribution.

```python
sns.countplot(x='Attrition', data=df, palette="mako")
plt.title("Attrition Distribution")
plt.show()
```

- Attribution By department:

```python
sns.countplot(x='Department', hue='Attrition', data=df, palette="crest")
plt.title("Attrition by Department")
plt.xticks(rotation=45)
plt.show()
```

- Percent Salary Hike vs Performance Rating

```python
sns.lineplot(x='PerformanceRating', y='PercentSalaryHike', data=df)
plt.title("Percent Salary Hike vs Performance Rating")
plt.show()
```

- Age group by attribution

```python
sns.countplot(x='AgeGroup', hue='Attrition', data=df)
plt.title("Attrition by Age Group")
plt.show()
```

```python
sns.count(x='YearsAtCompany', y='MonthlyIncome', hue='Attrition', data=df)
plt.title("Monthly Income vs Years at Company")
plt.show()
```

```python
sns.countplot(x='WorkLifeBalance', hue='Attrition', data=df)
plt.title("Work-Life Balance vs Attrition")
plt.show()
```
