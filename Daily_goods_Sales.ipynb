- **Convert Dates into datetime format**

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Production Date'] = pd.to_datetime(df['Production Date'], dayfirst=True)
df['Expiration Date'] = pd.to_datetime(df['Expiration Date'], dayfirst=True)
```

- **Product Performance Analysis**

```python
product_sales = df.groupby('Product Name')['Approx. Total Revenue(INR)'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
product_sales.plot(kind='bar', color = "green")
plt.title('Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Revenue (INR)')
plt.show()
```

Using this graph we can see the revenue generated by each dairy product. According to this chart the revenue of the curd is highest and the icecream revenue is a lowest.

- **Top Selling Products**

```python
top_products = df.groupby('Product Name')['Quantity Sold (liters/kg)'].sum().sort_values(ascending=False)
top_products.plot(kind='bar', figsize=(10,5), color = "Green")
plt.title('Quantity Sold by Product')
plt.show()
```

We can see that a the highest sold product is a curd and lowest sold product is a ghee.

- **Customer Preference Analysis**

```python
customer_pref = df.groupby('Product Name')['Quantity Sold (liters/kg)'].sum()

customer_pref.plot(kind='pie', autopct='%1.1f%%')
plt.title('Customer Preference by Product')
plt.ylabel('')
plt.show()
```

In Pie char we see that the customer's preference distribution across the products.

The Top three prefered products are:
1. Curd
2. Lassi
3. Milk and Butter

- **Sales Channel Analysis**

```python
channel_sales = df.groupby('Sales Channel')['Approx. Total Revenue(INR)'].sum()

channel_sales.plot(kind='bar', color = "Green")
plt.title('Revenue by Sales Channel')
plt.show()
```

By this comparison the highest revenue is retail and lowest revenue from the online sales.

- **Distribution Analysis**

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='Sales Channel', color = "green")
plt.title('Distribution of Sales Channels')
plt.show()
```

This shows distribution of sales.

- **Brand Performance**

```python
brand_sales = df.groupby('Brand')['Approx. Total Revenue(INR)'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,5))
brand_sales.plot(kind='bar', color="Green")
plt.title('Revenue by Brand')
plt.show()
```

Revenue of the Amul is highest and the lowest revenue from Britania Industries.

- **Revenue Trend**

```python
df['Year'] = df['Date'].dt.year

year_sales = df.groupby('Year')['Approx. Total Revenue(INR)'].sum()

year_sales.plot(marker='o')
plt.title('Year-wise Revenue Trend')
plt.ylabel('Revenue')
plt.show()
```

- **Inventory**

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Quantity in Stock (liters/kg)'], color = "Green")
plt.title('Inventory Distribution')
plt.show()
```

This shows the stock availability of product.

- **Correlation**

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

This shows the realtion in between the numerical values.
