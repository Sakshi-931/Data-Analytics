#### Plotting a bar chart for Gender and it's count

```python
ax = sns.countplot(x = 'Gender', data = df, hue = 'Gender', palette = ['orange','blue'], legend = False)

for bars in ax.containers:
    ax.bar_label(bars)
```

#### Plotting a bar chart for gender vs total amount

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by ='Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

#### Age

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')

for bars in ax.containers: 
    ax.bar_label(bars)
```

#### Total Amount VS Age Group

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age, hue = 'Age Group',palette = ['#0c3547','#10656d','#598f91','#93b071','#ede2cc','#edae93','#dd6670'])
```

#### State

#### Total Number of orders from top 10 states

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(16,6)})
sns.barplot(x = 'State', y = 'Orders',data = sales_state, hue = 'State',palette = ['#512889','#642e90','#773398','#8a399f','#9e3fa6','#b145ae','#c44ab5','#d750bd','#ea56c4','#ff5ccc'])
```

#### total amount/sales from top 10 states

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(10,7)})
sns.barplot(x = 'State', y = 'Amount',data = sales_state, hue = 'State',palette = ['#2a3e4b','#4cbbb3','#dcdfd8','#8f9c9d','#3d6f6d','#5a5f4e','#70d0cc','#428c7c','#44546c','#248c84'])
plt.xticks(rotation = 45)
plt.show() 
```

#### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Marital_Status', palette = ['#02876f','#f07801'])
sns.set(rc = {'figure.figsize':(7,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
marital_status = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount',data = marital_status, hue = 'Gender', palette = ['#ca0343','#234388'])
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

#### Occupation

```python
sns.set(rc = {'figure.figsize':(10,7)})
ax = sns.countplot(x = 'Occupation', data = df, hue = 'Occupation',palette = 'flare')

for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation = 45)
plt.show()  
```

```python
occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.set(rc = {'figure.figsize':(10,7)})
sns.barplot(x = 'Occupation', y = 'Amount',data = occupation, hue = 'Occupation',palette = 'flare')
plt.xticks(rotation = 45)
plt.show()
```

#### Product Categry

```python
sns.set(rc = {'figure.figsize':(12,7)})
ax = sns.countplot(x = 'Product_Category', data = df, hue = 'Product_Category',palette = 'crest')
plt.xticks(rotation = 45)

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
products = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(data = products, x = 'Product_Category',y = 'Amount', hue = 'Product_Category', palette = 'crest')
plt.xticks(rotation = 45)
plt.show()
```

```python
products = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.barplot(data = products, x = 'Product_ID',y = 'Orders', hue = 'Product_ID', palette = 'crest')
plt.xticks(rotation = 45)
plt.show()
```
